```python
import pandas as pd 
import numpy as np 
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
train_normal = '../../data/chest_xray/train/NORMAL'
train_pneumonia = '../../data/chest_xray/train/PNEUMONIA'
print(f"Normal images: {len(os.listdir(train_normal))}")
print(f"Pneumonia images: {len(os.listdir(train_pneumonia))}")
```

```python
train_normal
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
for i in range(3):
    img = Image.open(os.path.join(train_normal, os.listdir(train_normal)[i]))
    axes[0, i].imshow(img, cmap='gray')
    axes[0, i].set_title('Normal')
    
    img = Image.open(os.path.join(train_pneumonia, os.listdir(train_pneumonia)[i]))
    axes[1, i].imshow(img, cmap='gray')
    axes[1, i].set_title('Pneumonia')
plt.show()
```
